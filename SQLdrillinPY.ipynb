What is the mean movie length? Calculate this by defining a function.

```
def mean(nums):
    sum1 = 0
    for num in nums:
        sum1 += num
    length = len(nums)
    return sum1/length
```

```
sql = """
SELECT length 
FROM film;
"""

results = engine.execute(sql)
engine.dispose()
rows = results.fetchall()

# df_movlen = pd.read_sql_query(sql,con=engine)
results = [x['length'] for x in rows]
print(mean(results))
```

What is the median movie length? Calculate this by defining a function.

```
def median(nums):
    nums.sort()
    count = len(nums)
    if count%2 == 1:
        mid = math.floor(count / 2)
        median = nums[mid]
        return median
    else:
        mid1 = math.floor(count / 2) 
        a = nums[mid1] 
        b = nums[mid1 + 1]
        ans = (a + b)/2
        return ans

print (median(results))
```

Calculate the standard deviation of the movie lengths. Calculate this by defining a function.



```
def std_dev(nums):
    N = len(nums)
    sum = 0
    mean1 = mean(nums)
    for num in nums:
        diff = num - mean1
        sqrd = diff**2
        sum +=sqrd
    return math.sqrt(sum/ N-1)

print(std_dev(results))  
```
